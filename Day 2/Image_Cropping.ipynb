```python
# Hands-on Image Processing using OpenCV by Dr.Sushitha Susan Joseph
# Image Cropping using the image zebra.jpg
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("batman 144p.jpg")
```

```python
cv2_imshow(img)
img.shape
```

```python
# img[row_start:row_end, column_start:column_end]
# Height = 422 pixels
# Width  = 406 pixels

crop = img[30:400, 230:400]
cv2_imshow(crop)
```

```python
print(f"Original image shape: {img.shape}")

center_crop = img[30:400, 118:288]
cv2_imshow(center_crop)
print(f"Center crop shape: {center_crop.shape}")
```

```python
left_crop = img[30:400, 0:170]
cv2_imshow(left_crop)
print(f"Left crop shape: {left_crop.shape}")
```
