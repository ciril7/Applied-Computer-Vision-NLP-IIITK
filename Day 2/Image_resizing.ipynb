```python
img = cv2.imread("shambu.jpg")
```

```python
cv2_imshow(img)
```

```python
print(img.shape)
```

```python
# Resize image to 300 × 300 pixels
resized_img = cv2.resize(img, (300, 300))

cv2_imshow(resized_img)
```

```python
print(resized_img.shape)
```

```python
# Reduce to half of the original dimension
# Get original dimensions
height, width = img.shape[:2]

# Resize to half size, The // operator performs integer division.
half_img = cv2.resize(img, (width//2, height//2))

cv2_imshow(half_img)
```

```python
print(half_img.shape)
```
