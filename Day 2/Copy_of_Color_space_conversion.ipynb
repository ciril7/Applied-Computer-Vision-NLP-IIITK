```python
# Hands-on Image Processing using OpenCV by Dr.Sushitha Susan Joseph
# Image color space conversion using the image bird.jpg
# OpenCV stores color images using three channels: Blue, Green, and Red (BGR).
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("pele.jpg")
```

```python
cv2_imshow(img)
img.shape
```

```python
# Convert color image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
```

```python
print(gray.shape)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

cv2_imshow(hsv)
```

```python
# Thresholding
# Convert grayscale image to a binary image using threshold value 127
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

print(ret)
cv2_imshow(thresh)
```

```python
#Instead of manually choosing 127, can the computer automatically choose a threshold?
_, otsu = cv2.threshold(gray, 0, 255,
                        cv2.THRESH_BINARY + cv2.THRESH_OTSU)

cv2_imshow(otsu)
```

```python
# Adaptive thresholding
# Instead of one threshold for the whole image, OpenCV computes a local threshold for each region.
adaptive = cv2.adaptiveThreshold(
    gray,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)

cv2_imshow(adaptive)
```
