```python
# Hands-on Image Processing using OpenCV by Dr.Sushitha Susan Joseph
# Edge detection using the image coin.jpg
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("coin.png")
```

```python
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
```

```python
# Smooth the image and reduce noise using Gaussian Blur
blur = cv2.GaussianBlur(gray, (5,5), 0)

cv2_imshow(blur)
```

```python
# Smooth the image and reduce noise using Gaussian Blur with a small kernel size
blur1 = cv2.GaussianBlur(gray, (3,3), 0)

cv2_imshow(blur1)
```

```python
# Apply Canny Edge Detection on blurred image
# Pixels with strong intensity changes are likely to be edges
edges = cv2.Canny(blur, 100, 200)

cv2_imshow(edges)
```

```python
# Without explicit blur apply canny edge detection
edges1 = cv2.Canny(gray, 100, 200)
cv2_imshow(edges1)
```
