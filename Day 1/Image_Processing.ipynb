```python
import numpy as np
import cv2
from google.colab import drive
from google.colab.patches import cv2_imshow
drive.mount('/content/drive')

kernel = np.ones((5,5), np.uint8)
print (kernel)
img = cv2.imread('/content/drive/My Drive/IIITK files/fifa cover.jpg')
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # Fixed: BGR2GRAY
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Fixed: cvt.Color to cvtColor and space in COLOR
ImgBlur =cv2.GaussianBlur(imgGray, (7,7),0)
imgCanny = cv2.Canny(ImgBlur, 100, 200) # Fixed: imgBlur to ImgBlur
ImgDilation =cv2.dilate(imgCanny,kernel, iterations=10)
imgEroded = cv2.erode(ImgDilation, kernel, iterations=2)
print("rgbimage",imgRGB.shape) # Fixed: LmgRGB to imgRGB
cv2_imshow(img) # Fixed: cv2.imshow to cv2_imshow
cv2_imshow(imgGray) # Fixed: cv2.imshow to cv2_imshow
cv2_imshow(ImgBlur) # Fixed: cv2.imshow to cv2_imshow
cv2_imshow(imgCanny) # Fixed: cv2.imshow to cv2_imshow
cv2_imshow(ImgDilation) # Fixed: cv2.imshow to cv2_imshow
cv2_imshow(imgEroded)
```
